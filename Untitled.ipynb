```python
# Assume 'df' is the concatenated DataFrame

# Split the concatenated DataFrame into train and test
train = df.iloc[:train_data.shape[0]]
test = df.iloc[train_data.shape[0]:]
```

```python
#input split
X=train.drop(columns='IS_HCP',axis=1)
y=train['IS_HCP']
```

```python
# Importing the necessary library
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape,y_train.shape
```

```python
train.columns[df.isnull().any()]
```

```python
# Dropping Rows with Missing Values in the 'IS_HCP' Column

train = train.dropna(subset=['IS_HCP'])
```

```python
train.isnull().sum()
```

```python
# Importing the necessary library
from sklearn.ensemble import RandomForestClassifier

# Creating a RandomForestClassifier model
model = RandomForestClassifier()

# Fitting the model to the training data
model.fit(X_train, y_train)
```

```python
# Making Predictions on Test Data

y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Calculating Accuracy Score

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
